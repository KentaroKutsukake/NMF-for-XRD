```python
import numpy as np
from matplotlib import pyplot as plt
from pylab import rcParams
from sklearn.decomposition import NMF
%matplotlib inline
```

### Load data 
##### 961 XRD patterns with reduced pixel resolution (33 x 33 pixels) were reshaped and stacked into a numpy array (961, 1089).  


```python
Processed_pattern = np.load("Processed_pattern.npy")
Processed_pattern.shape
```

### Figure 4. Feature patterns obtained by NMF. 

```python
n_dim = 2

model = NMF(n_components=n_dim)
X_embedded = model.fit_transform(Processed_pattern)
X_comp = model.components_

plt.figure(figsize=(1.5*n_dim,2), dpi=150)
for i in range(n_dim):
    plt.subplot(1, n_dim, i+1)
    plt.imshow(X_comp[i].reshape(33,33))
    plt.tick_params(labelbottom=False, bottom=False,labelleft=False, left=False)

plt.show()
```

### Figure 5. Spatial mapping images of differences of the coefficients.

```python
n_dim = 4

model = NMF(n_components=n_dim)
X_embedded = model.fit_transform(Processed_pattern)
X_comp = model.components_

rcParams['figure.figsize'] = 4, 3.5
xp = np.linspace(0,150,31)
yp = np.linspace(0,150,31)
plt_y, plt_x = np.meshgrid(xp,yp)
plt.scatter(plt_x.reshape(-1), plt_y.reshape(-1), marker="s", c=X_embedded[:,2]-X_embedded[:,0], s=110)
plt.xlabel("Position(m)")
plt.ylabel("Position(m)")
plt.ylim(0,150)
plt.xlim(0,150)
plt.xticks([0, 50,100,150])
plt.yticks([0, 50,100,150])
plt.colorbar()
plt.show()
```
